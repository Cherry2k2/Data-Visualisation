```python
df = pd.read_table(r'C:\Users\Msc 1\Desktop\MDSC-106\diabetes.tab.txt')
df = df.rename(columns={'AGE': 'age', 'SEX': 'sex', 'BMI':'bmi', 'BP':'bp','S1':'s1', 'S2':'s2', 'S3':'s3', 'S4':'s4', 'S5':'s5', 'S6':'s6', 'Y':'target'})
df.head()
```

```python
plt.figure(figsize=(20, 7))

plt.subplot(1, 4, 1)
sns.histplot(df['age'], kde = True,color='r')
plt.title('Histogram')

# scatterplot
plt.subplot(1, 4, 2)
plt.scatter(df['age'],df['bp'],color = 'g')
plt.title('Scatterplot')


# boxplot
plt.subplot(1, 4, 3)
sns.boxplot(y=df['target'],color = 'b')
plt.title('Boxplot')

# barplot
plt.subplot(1, 4, 4)
sns.barplot(x='target', y = 'bmi', data = df)   
plt.title('Barplot')

plt.show()
```

# Stacked barplot in matplotlib

```python
import pandas as pd
from sklearn import datasets as data
df = data.load_diabetes(as_frame=True)
df['data']
```

```python
import matplotlib.pyplot as plt
```

```python
x=['A', 'B', 'C', 'D']
y1=[40, 20, 30, 40]
y2=[10, 20, 10, 50]
plt.bar(x,y1)
plt.bar(x,y2, bottom=y1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 30, 40])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)
plt.bar(x, y3, bottom=y1+y2)
plt.bar(x, y4, bottom=y1+y2+y3)
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["round1", "round2", "round3", "round4"])
plt.title("Scores by teams in 4 rounds")
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame([['A', 10, 20, 30, 40], ['B', 20, 25, 15, 25], ['C', 12, 15, 19, 6], ['D', 10, 29, 13, 19]], columns=['Team', 'Round1', 'Round2', 'Round3', 'Round4'])
print(df)
df.plot(x='Team', kind='bar', stacked='True', title='Stacked bargraph by dataframe')
plt.show()
```
